```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取Excel数据
df = pd.read_excel("4. 下一步研究的景点列表（79个免费4.8分以上）.xlsx", engine="openpyxl")

# 查看数据列名
print("数据列名：", df.columns.tolist())

# 按“分类”列统计数量，按数量降序排列
type_counts = df["分类"].value_counts().sort_values(ascending=False)

# 计算总数量和各类型占比
total = type_counts.sum()
type_percent = (type_counts / total * 100).round(1)  # 占比保留1位小数
```

```python
# 设置中文字体（解决中文显示不出的问题）
plt.rcParams["font.family"] = [ "Heiti TC", "PingFang SC"]
plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示问题

# 创建画布
fig, ax = plt.subplots(figsize=(12, 8))

# 绘制柱状图
bars = ax.bar(
    x=type_counts.index,  # X轴：景点分类
    height=type_counts.values,  # Y轴：数量
    color="#409EFF",
    alpha=0.8,
    edgecolor="#1E88E5",
    linewidth=1.5,
)
```
