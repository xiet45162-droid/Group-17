```python
import pandas as pd
file_path = "ctrip_sights_300pages_20251122_144901.csv"
df = pd.read_csv(file_path)
# 1. 读取Excel文件
# 请将此处的文件路径替换为你的Excel文件实际路径
```

```python

split_columns = df['location'].str.split(' · ', expand=True)
# 给新列命名
df['城市'] = split_columns[0]  # 第一部分作为城市
df['景点'] = split_columns[1]  # 第二部分作为景点
# 2. 拆分location列
# 使用" · "作为分隔符，expand=True表示拆分成多列
# 拆分后会得到两列，分别是城市和景点
```

```python
# 3. 处理没有分隔符的情况
# 将景点列中的NaN值替换为空字符串
df['景点'] = df['景点'].fillna('')
```

```python
# 4. 可选：如果需要删除原始的location列
df = df.drop('location', axis=1)
```

```python
# 5. 保存处理后的结果到新的Excel文件
output_file = "processed_file.xlsx"
df.to_excel(output_file, index=False)

print(f"处理完成，结果已保存到 {output_file}")
```
