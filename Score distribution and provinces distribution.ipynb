```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 读取数据，使用GBK编码解决中文编码问题
file_path = r'D:\2025 at CityU\semester A\5507 Soc Meida data\final\5507数据\3000景点清洗后数据.csv'
try:
    df = pd.read_csv(file_path, encoding='gbk')
except:
    try:
        df = pd.read_csv(file_path, encoding='utf-8')
    except:
        df = pd.read_csv(file_path, encoding='latin-1')

print(f"总记录数: {len(df)}")

# 检查'free'列的唯一值
print("\n'free'列的唯一值:")
print(df['free'].unique())

# 检查列的数据类型
print(f"\n'free'列的数据类型: {df['free'].dtype}")

# 查看样本数据
print("\n'free'列样本数据:")
print(df[['name', 'free']].head(10))

# 清理'free'列数据（去除空格，统一大小写）
df['free_cleaned'] = df['free'].astype(str).str.strip().str.upper()

# 筛选免费景点
free_attractions = df[df['free_cleaned'] == 'TRUE']

print(f"\n免费景点数量: {len(free_attractions)}")

if len(free_attractions) > 0:
    # 处理评分数据，确保是数值类型
    free_attractions['rating'] = pd.to_numeric(free_attractions['rating'], errors='coerce')
    
    # 移除评分缺失的记录
    free_attractions = free_attractions.dropna(subset=['rating'])
    
    # 过滤掉评分显示为5.0以上的干扰项
    free_attractions = free_attractions[free_attractions['rating'] <= 5.0]
    
    print(f"处理后免费景点数量: {len(free_attractions)}")
    
    if len(free_attractions) > 0:
        # 查看评分统计信息
        print("免费景点评分统计:")
        print(free_attractions['rating'].describe())
        
        # 打印前5行样本数据
        print("\n样本数据（前5行）:")
        print(free_attractions[['name', 'rating']].head())
        
        # 查看评分的唯一值
        print("\n发现的唯一评分值:")
        print(sorted(free_attractions['rating'].unique()))
        
        # 创建评分分类
        # 首先，将所有评分四舍五入到1位小数
        free_attractions['rounded_rating'] = free_attractions['rating'].round(1)
        
        # 将评分≤4.0的归为一类
        def create_category(rating):
            if rating <= 4.0:
                return '≤4.0'
            else:
                return f'{rating:.1f}'
        
        free_attractions['rating_category'] = free_attractions['rating'].apply(create_category)
        
        # 统计每个分类的景点数量
        rating_counts = free_attractions['rating_category'].value_counts()
        
        print("\n原始评分数量统计:")
        print(rating_counts)
        
        # 创建完整的分类列表（按顺序）
        # 首先，创建从4.1到5.0的所有可能评分值
        all_categories = ['≤4.0']
        for i in range(1, 10):  # 4.1到4.9
            all_categories.append(f'4.{i}')
        all_categories.append('5.0')
        
        # 重新索引rating_counts以包含所有分类，缺失的用0填充
        rating_counts = rating_counts.reindex(all_categories, fill_value=0)
        
        print("\n最终评分分布:")
        for category, count in rating_counts.items():
            print(f"{category}: {count}个景点")
        
        # 创建折线图
        plt.figure(figsize=(14, 8))
        
        # 绘制数据
        x_positions = range(len(rating_counts))
        plt.plot(x_positions, rating_counts.values, marker='o', linewidth=2, markersize=8, color='#2E86AB')
        
        # 格式化图表
        plt.title('免费景点评分分布图', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('评分', fontsize=12)
        plt.ylabel('景点数量', fontsize=12)
        
        # 设置X轴刻度
        plt.xticks(x_positions, rating_counts.index, rotation=45)
        plt.grid(True, alpha=0.3)
        
        # 在点上方添加数值标签
        for i, (x, y) in enumerate(zip(x_positions, rating_counts.values)):
            plt.annotate(str(y), (x, y), textcoords="offset points", xytext=(0,10), 
                        ha='center', fontsize=9)
        
        plt.tight_layout()
        
        # 保存图表
        output_path = r'D:\2025 at CityU\semester A\5507 Soc Meida data\final\免费景点评分分布图.png'
        plt.savefig(output_path, dpi=300, bbox_inches='tight')
        plt.show()
        
        print(f"\n图表已保存至: {output_path}")
        
        # 显示详细分布
        print("\n免费景点详细评分分布:")
        for category in rating_counts.index:
            count = rating_counts[category]
            if count > 0:
                attractions_in_category = free_attractions[free_attractions['rating_category'] == category]
                print(f"\n{category}评分 ({count}个景点):")
                for _, attraction in attractions_in_category.iterrows():
                    print(f"  - {attraction['name']} ({attraction['城市']}): {attraction['rating']}")
        
        # 显示评分最高和最低的免费景点
        print("\n评分最高的5个免费景点:")
        top_free_attractions = free_attractions.nlargest(5, 'rating')[['name', '城市', 'rating']]
        print(top_free_attractions.to_string(index=False))
        
        print("\n评分最低的5个免费景点:")
        bottom_free_attractions = free_attractions.nsmallest(5, 'rating')[['name', '城市', 'rating']]
        print(bottom_free_attractions.to_string(index=False))
        
        # 额外统计信息
        print(f"\n分析的免费景点总数: {len(free_attractions)}")
        print(f"平均评分: {free_attractions['rating'].mean():.2f}")
        print(f"中位数评分: {free_attractions['rating'].median():.2f}")
        print(f"评分≤4.0的景点数量: {len(free_attractions[free_attractions['rating'] <= 4.0])}")
        
        # 计算各评分区间的百分比
        print("\n各评分区间百分比分布:")
        total_attractions = len(free_attractions)
        for category in rating_counts.index:
            count = rating_counts[category]
            percentage = (count / total_attractions) * 100
            print(f"{category}: {percentage:.1f}%")
    else:
        print("未找到具有有效评分的免费景点。")
else:
    print("未找到免费景点。请检查CSV文件中'free'列的值。")
    print("'free'列的所有唯一值:")
    print(df['free'].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 设置输出路径
output_path = r'D:\2025 at CityU\semester A\5507 Soc Meida data\final'

# 尝试不同的编码方式读取文件
def read_csv_with_encodings(file_path):
    encodings = ['utf-8', 'gbk', 'gb2312', 'latin1', 'cp1252']
    
    for encoding in encodings:
        try:
            print(f"尝试使用 {encoding} 编码读取文件...")
            df = pd.read_csv(file_path, encoding=encoding)
            print(f"成功使用 {encoding} 编码读取文件")
            return df
        except UnicodeDecodeError:
            continue
        except Exception as e:
            print(f"使用 {encoding} 编码时出现错误: {e}")
            continue
    
    # 如果所有编码都失败，尝试使用错误处理
    try:
        print("尝试使用 errors='ignore' 读取文件...")
        df = pd.read_csv(file_path, encoding='utf-8', errors='ignore')
        return df
    except Exception as e:
        print(f"所有编码方式都失败: {e}")
        return None

# 读取样本文件
file1_path = os.path.join(output_path, 'next step_with_province.csv')
df1 = read_csv_with_encodings(file1_path)

if df1 is None:
    print("无法读取第一个文件")
    exit()

# 读取3000个景点列表文件
file2_path = os.path.join(output_path, '3000 city.csv')
df2 = read_csv_with_encodings(file2_path)

if df2 is None:
    print("无法读取第二个文件")
    exit()

# 显示数据框的基本信息
print("\n第一个文件的基本信息:")
print(f"列名: {df1.columns.tolist()}")
print(f"行数: {len(df1)}")
print("\n前几行数据:")
print(df1.head())

print("\n第二个文件的基本信息:")
print(f"列名: {df2.columns.tolist()}")
print(f"行数: {len(df2)}")
print("\n前几行数据:")
print(df2.head())

# 检查并清理省份列名
# 对于第一个文件，省份列可能是 '所属省/直辖市/自治区/特别行政区'
# 对于第二个文件，省份列是 '省份'
province_col1 = '所属省/直辖市/自治区/特别行政区'
province_col2 = '省份'

# 检查列名是否存在
if province_col1 not in df1.columns:
    print(f"第一个文件中找不到列 '{province_col1}'")
    print(f"可用的列: {df1.columns.tolist()}")
    # 尝试找到包含'省'的列
    for col in df1.columns:
        if '省' in col:
            province_col1 = col
            print(f"使用列 '{province_col1}' 作为省份列")
            break

if province_col2 not in df2.columns:
    print(f"第二个文件中找不到列 '{province_col2}'")
    print(f"可用的列: {df2.columns.tolist()}")
    # 尝试找到包含'省'的列
    for col in df2.columns:
        if '省' in col:
            province_col2 = col
            print(f"使用列 '{province_col2}' 作为省份列")
            break

# 统计各省份景点数量
print("\n正在统计各省份景点数量...")

# 清理数据，移除空值
df1_clean = df1.dropna(subset=[province_col1])
df2_clean = df2.dropna(subset=[province_col2])

province_count1 = df1_clean[province_col1].value_counts()
province_count2 = df2_clean[province_col2].value_counts()

print(f"\n第一个文件中的省份统计:")
print(province_count1)
print(f"\n第二个文件中的省份统计:")
print(province_count2)

# 合并两个统计结果
all_provinces = list(set(province_count1.index) | set(province_count2.index))
all_provinces = [p for p in all_provinces if pd.notna(p)]  # 移除空值
all_provinces.sort()

print(f"\n总共找到 {len(all_provinces)} 个不同的省份/地区")

# 创建数据框用于绘图
plot_data = pd.DataFrame({
    'next_step_file': [province_count1.get(province, 0) for province in all_provinces],
    'city_3000_file': [province_count2.get(province, 0) for province in all_provinces]
}, index=all_provinces)


# 按降序排序
plot_data = plot_data.sort_values('city_3000_file', ascending=False)

# 创建图形
plt.figure(figsize=(16, 10))

# 创建分组柱状图（两个文件的对比）
bar_width = 0.35
x_pos = np.arange(len(plot_data.index))

bars1 = plt.bar(x_pos - bar_width/2, plot_data['next_step_file'], 
                bar_width, label='next_step_with_province.csv', alpha=0.8, color='skyblue')
bars2 = plt.bar(x_pos + bar_width/2, plot_data['city_3000_file'], 
                bar_width, label='3000_city.csv', alpha=0.8, color='lightcoral')

plt.xlabel('省份', fontsize=12)
plt.ylabel('景点数量', fontsize=12)
plt.title('各省份景点数量对比', fontsize=14, fontweight='bold')
plt.xticks(x_pos, plot_data.index, rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# 在柱子上添加数值标签
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'{int(height)}', ha='center', va='bottom', fontsize=8)

# 调整布局
plt.tight_layout()

# 保存图片
output_image_path = os.path.join(output_path, 'province_attraction_comparison.png')
plt.savefig(output_image_path, dpi=300, bbox_inches='tight')
print(f"\n图表已保存至: {output_image_path}")

# 显示图表
plt.show()

# 输出统计摘要
print("\n统计摘要:")
print(f"文件1中的省份数量: {len(province_count1)}")
print(f"文件2中的省份数量: {len(province_count2)}")

# 保存详细数据到CSV
detailed_data_path = os.path.join(output_path, 'province_attraction_statistics.csv')
plot_data.to_csv(detailed_data_path, encoding='utf-8-sig')
print(f"\n详细数据已保存至: {detailed_data_path}")

# 打印文件基本信息
print(f"\n文件统计:")
print(f"next_step_with_province.csv: {len(df1)} 行, {len(province_count1)} 个省份")
print(f"3000_city.csv: {len(df2)} 行, {len(province_count2)} 个省份")
```
